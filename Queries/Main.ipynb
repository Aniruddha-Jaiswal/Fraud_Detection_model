```python
df=df.drop(['step','type','nameOrig','newbalanceOrig','nameDest','oldbalanceDest','isFlaggedFraud'],axis=1)
```

```python
#add 2 colums named condition 1 and condition 2
#Condition 1:100% of transactions are fraud whose “ amount== oldbalanceOrg ”. 
#We can find 8034 out of 8213 fraud transaction with these observation.
df['cond1']=(df['amount']==df['oldbalanceOrg']).astype(int)
```

```python
#	Condition 2: 100% of transactions are fraud whose ‘newbalanceDest==0’ and “amount ==10000000”.
#We can find further 145 fraud transaction with these condition (145 out of remaining 179).
df['cond2']=((df['amount']==10000000.00) & (df['newbalanceDest']==0)).astype(int)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X=df[['amount','oldbalanceOrg','newbalanceDest','cond1','cond2']]
Y=df['isFraud']
```

```python
X_train,X_test,Y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# we are going to use Random Forest Classifier as it combines predictions of multipke trees and reduce overfitting chances.
```

```python
model=RandomForestClassifier(random_state=42)
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
y_pred=model.predict(X_test)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred))
```

```python
print('Classification:',classification_report(y_test,y_pred))
```

```python
print('Confusion Matrix:',confusion_matrix(y_test,y_pred))
```

```python
# overall model performs well and have accuracy score of 99%.
```
